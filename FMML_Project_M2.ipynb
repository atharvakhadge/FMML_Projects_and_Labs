````python
## we will be comparing rest of the parameters/columns present in the data with respect to precence or absece of heart disease
data['target'] = data.target.replace({1: "Disease", 0: "No_disease"})
data['sex'] = data.sex.replace({1: "Male", 0: "Female"})
data['cp'] = data.cp.replace({1: "typical_angina",
                          2: "atypical_angina",
                          3:"non-anginal pain",
                          4: "asymtomatic"})
data['exang'] = data.exang.replace({1: "Yes", 0: "No"})
data['slope'] = data.cp.replace({1: "upsloping",
                          2: "flat",
                          3:"downsloping"})
data['thal'] = data.thal.replace({1: "fixed_defect", 2: "reversable_defect", 3:"normal"})
````

````python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
````

````python
#to do
# Create a pie chart of the 'Tea' column also write your observation form the plot
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('star_nutri_expanded.csv')

# Filter for tea-related beverages (assuming they are categorized under 'Tea' or similar)
tea_data = data[data['Beverage_category'].str.contains('Tea', case=False, na=False)]

# Count occurrences of each type of tea
tea_counts = tea_data['Beverage'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(tea_counts, labels=tea_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Tea Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

````

````python
## to do
# perform pca on the data and plot the explained variace ratio, what is the optimal number of principal components in this case ?
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select numerical columns for PCA
numerical_data = data.select_dtypes(include=['float64', 'int64']).dropna()

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numerical_data)

# Perform PCA
pca = PCA()
pca.fit(scaled_data)

# Plot Explained Variance Ratio
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.title('Explained Variance Ratio by Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, len(pca.explained_variance_ratio_) + 1))
plt.grid()
plt.show()

# Determine Optimal Components
cumulative_variance = pca.explained_variance_ratio_.cumsum()
optimal_components = (cumulative_variance < 0.95).sum()  # For example, to capture 95% variance
print(f'Optimal number of principal components to capture 95% variance: {optimal_components}')
````
